```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Geopandas kütüphanesi ile Türkiye haritasını yükleyin
turkey = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
turkey = turkey[turkey['name'] == 'Turkey']

# Haritayı çizin
fig, ax = plt.subplots(figsize=(10, 10))
turkey.plot(ax=ax, color='lightblue', edgecolor='black')

# Grafik başlığını ve eksen etiketlerini ekleyin
plt.title('Türkiye Haritası')
plt.xlabel('Boylam')
plt.ylabel('Enlem')

# Haritayı göster
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Türkiye'nin sınırlarını içeren dünya haritasını yükleyin
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Türkiye sınırlarını seçin
turkey = world[world['name'] == 'Turkey']

# Şehir merkezlerini içeren veri kümesini yükleyin
cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))

# Şehir merkezlerini Türkiye sınırları içine alın
turkey_cities = gpd.sjoin(cities, turkey, how='inner', op='intersects')

# Haritayı çizin
fig, ax = plt.subplots(figsize=(10, 10))

# Türkiye sınırlarını çizin
turkey.plot(ax=ax, color='lightblue', edgecolor='black')

# Şehir merkezlerini nokta olarak çizin
turkey_cities.plot(ax=ax, color='red', markersize=50, marker='o', label='Şehir Merkezleri')

# Grafik başlığını ve eksen etiketlerini ekleyin
plt.title('Türkiye Haritası - Şehir Merkezleri')
plt.xlabel('Boylam')
plt.ylabel('Enlem')

# Lejantı göster
plt.legend()

# Haritayı göster
plt.show()
```
